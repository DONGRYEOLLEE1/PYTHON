```python
df3 = pd.read_csv('examples/ex5.csv')
# NA값 삭제, INDEX 생성
df3.to_csv('examples/my_out_of_ex5.csv', na_rep='ToBeReplaced')            # na_rep='ToBeReplaced'  =>  NA값을 'ToBeReplaced'로 대체
```

```python
df3 = pd.read_csv('examples/ex5.csv')
# INDEX, HEADER 값 읽지 않기, 원하는 COLUMN 값만 읽기
df3.to_csv('examples/my_out_of_ex5.csv', na_rep='ToBeReplaced', index = False, columns = ['a', 'b', 'c'])
```

```python
### Web Crawling

from bs4 import BeautifulSoup
import requests
```

```python
## 기상청 홈페이지 웹크롤링
source = requests.get("https://www.weather.go.kr/weather/observation/currentweather.jsp")
soup = BeautifulSoup(source.content, "html.parser")
table = soup.find('table', {"class":"table_develop3"})
table
```

```python
## 웹크롤링 교육용 홈페이지 
url = "http://www.webscrapingfordatascience.com/basichttp/"
r = requests.get(url)
r.text
```

```python
# Find to "Season1 Ep1. Title"
url = "https://en.wikipedia.org/w/index.php?title=List_of_Game_of_Thrones_episodes&oldid=802553687"
source1 = requests.get(url)
soup1 = BeautifulSoup(source1.content, "html.parser")
table1 = soup1.find('table', {"class":"wikitable plainrowheaders wikiepisodetable"})
td1 = table1.find('td', {"class":"summary"})
td1.text.replace('"', "")
```
