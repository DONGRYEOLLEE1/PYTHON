```python
import pandas as pd
import numpy as np
import mglearn 
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
plt.figure(figsize=(8,8))
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
```

```python
# PREDICTION
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
# ACCURACY 구해보기
clf.score(X_test, y_test)
### 100% 도출되면 OVERFITTING 의심해봐야함 !
```

```python
# 결정 경계 Decision Boundary
fig, axes = plt.subplots(1, 3, figsize=(10,3))
for n, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
axes[0].legend(loc=3)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer['DESCR']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=66)
train_res = []
test_res = []
for n in range(1, 11):
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
    train_res.append(clf.score(X_train, y_train))
    test_res.append(clf.score(X_test, y_test))

plt.plot(range(1,11), train_res, label = '훈련 정확도')
plt.plot(range(1,11), test_res, label = '테스트 정확도')
plt.ylabel("정확도")
plt.xlabel("n")
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = KNeighborsRegressor(n_neighbors=9).fit(X_train, y_train)
print(X_test.reshape(-1,10))
print(reg.predict(X_test))
```

```python
reg.score(X_test, y_test)
# n = 1 , 35.2
# n = 3 , 83.4
# n = 9 , 65.4
```
