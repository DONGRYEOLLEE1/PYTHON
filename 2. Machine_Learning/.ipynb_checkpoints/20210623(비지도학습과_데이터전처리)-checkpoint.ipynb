```python
# 지도학습에서의 데이터전처리 효과

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
from sklearn.svm import SVC
svm = SVC(gamma = 'auto').fit(X_train, y_train)
print(svm.score(X_test, y_test))
```

```python
# SVC 모델을 학습시키기 전에 MinMaxScaler을 사용해 데이터 스케일을 조정해보자 (MinMaxScaler)

scaler = MinMaxScaler().fit(X_train, y_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)

svm = SVC(gamma = 'auto').fit(X_train_s, y_train)
print(svm.score(X_test_s, y_test))
```

```python
# 이번엔 평균 0, 분산 1을 갖도록 스케일을 조정해보자 (StandardScaler)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train, y_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)

svm = SVC(gamma = 'auto').fit(X_train_s, y_train)
print(svm.score(X_test_s, y_test))
```

```python
## 원형이 아닌 클러스터를 구분하지 못하는 k평균 알고리즘
import numpy as np
import matplotlib.pyplot as plt
X, y = make_blobs(random_state=170, n_samples = 600)
rng = np.random.RandomState(74)

# 데이터가 길게 늘어지도록 변경
transformation = rng.normal(size=(2,2))
X = np.dot(X, transformation)

# 세 개의 클러스터로 데이터에 KMeans 알고리즘을 적용
kmeans = KMeans(n_clusters=3).fit(X)
y_pred = kmeans.predict(X)

# 클러스터 할당, 클러스터 중심을 나타내기
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], [0, 1, 2], 
                         markers='^', markeredgewidth = 2)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
```

```python
# Chapter 12 군집 분석
import pandas as pd
import math
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
### 데이터 정제하기

# 오류데이터 정제 
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
df = df[df['CustomerID'].notnull()]

# 'CustomerID' 자료형을 정수형으로 변환
df['CustomerID'] = df['CustomerID'].astype(int)

# 확인
df.info()
```

```python
# 중복 레코드 제거
df.drop_duplicates(inplace = True)
df.info()
```

```python
df['SaleAmount'] = df['UnitPrice'] * df['Quantity']
df.head()
```

```python
work_table = {
    'InvoiceNo' : 'count',
    'SaleAmount' : 'sum',
    'InvoiceDate' : 'max'
}
customer_df = df.groupby('CustomerID').agg(work_table)
customer_df = customer_df.reset_index()
customer_df
```

```python
customer_df = customer_df.rename(columns = {'InvoiceNo' : 'Freq', 'InvoiceDate' : 'ElapsedDays'})
customer_df.head()
```

```python
import datetime
print(datetime.datetime(2011,12,10))
customer_df['ElapsedDays'] = datetime.datetime(2011,12,10) - customer_df['ElapsedDays']
customer_df.head()
```

```python
customer_df['ElapsedDays'] = customer_df['ElapsedDays'].apply(lambda x: x.days+1)
customer_df.head()
```

```python
# 데이터 분포 조정하기
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9,9))
ax.boxplot([customer_df['Freq'], customer_df['SaleAmount'], customer_df['ElapsedDays']], sym = 'bo')
plt.xticks([1, 2, 3], ['Freq', 'SaleAmount', 'ElapsedDays'])
plt.show()
```

```python
# X_faetures를 정규 분포로 스케일링
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
X = customer_df[['Freq_log', 'SaleAmount_log', 'ElapsedDays_log']].values
X
```

```python
from sklearn.preprocessing import StandardScaler
X_s = StandardScaler().fit_transform(X)
```

```python
# 엘보 방법으로 클러스터 개수 k 선택

distortions = []

for i in range(1, 11):
    kmeans_i = KMeans(n_clusters = i, random_state=0).fit(X_s)
    distortions.append(kmeans_i.inertia_)   # inertia_ : 왜곡 값
    
plt.plot(range(1,11), distortions, marker = 'o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
### 엘보(꺾이는 지점)라고 할 수 있는 지점이 3 또는 4,5가 될 수 있을 것 같음.
```

```python
kmeans_k3 = KMeans(n_clusters = 3, random_state=0)
y_ = kmeans_k3.fit_predict(X_s)
customer_df['ClusterLabel'] = y_
```

```python
# 클러스터의 비중과 데이터 분포를 차트로 시각화
from matplotlib import cm

def silhouetteViz(n_cluster, X):
    kmeans = KMeans(n_clusters = n_cluster, random_state = 0)
    y_ = kmeans.fit_predict(X)
    
    silhouette_values = silhouette_samples(X, y_, metric = 'euclidean')
    
    y_ax_lower, y_ax_upper = 0, 0
    y_ticks = []
    
    for c in range(n_cluster):
        c_silhouettes = silhouette_values[y_ == c]
        c_silhouettes.sort()
        y_ax_upper += len(c_silhouettes)
        color = cm.jet(float(c) / n_cluster)
        plt. barh(range(y_ax_lower, y_ax_upper), c_silhouettes, height = 1.0, edgecolor = 'none', color = color)
        y_ticks.append((y_ax_lower + y_ax_upper) / 2.)
        y_ax_lower += len(c_silhouettes)
        
    silhouette_avg = np.mean(silhouette_values)
    plt.axvline(silhouette_avg, color = 'red', linestyle = '--')
    plt.title('Number of Cluster : '+ str(n_cluster) + '\n' +
              'Silhouette Score : '+ str(round(silhouette_avg, 3)))
    plt.yticks(y_ticks, range(n_cluster))
    plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.tight_layout()
    plt.show()
```

```python
def clusterScatter(n_cluster, X):
    c_colors=[]
    kmeans = KMeans(n_clusters = n_cluster, random_state=0)
    y_ = kmeans.fit_predict(X)
    
    for i in range(n_cluster):
        c_color = cm.jet(float(i) / n_cluster)
        c_colors.append(c_color)
        # 클러스터의 데이터 분포를 동그라미로 시각화
        plt.scatter(X[y_ == i, 0], 
                    X[y_ == i, 1], 
                    marker = 'o', color = c_color, 
                    edgecolor = 'black', s = 50, 
                    label = 'cluster ' + str(i))
        # 각 클러스터의 중심점을 삼각형으로 표시
    for i in range(n_cluster):
        plt.scatter(kmeans.cluster_centers_[i, 0], 
                    kmeans.cluster_centers_[i, 1], 
                    marker = '^', color = c_colors[i], 
                    edgecolor = 'w', s = 200)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
silhouetteViz(3, X)
silhouetteViz(4, X)   # 실루엣계수가 가장 높은 것은 클러스터 4개인 경우로 0.31
silhouetteViz(5, X)
silhouetteViz(6, X)
```

```python
clusterScatter(3, X)
clusterScatter(4, X)
clusterScatter(5, X)
clusterScatter(6, X)
```
