```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import QuantileTransformer, StandardScaler, PowerTransformer
```

```python
data = pd.read_excel('data/Dry_Bean_Dataset.xlsx')
data[:5]
```

```python
X = data.drop(['Class'],axis=1)
y = data['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
print(X_train.shape, X_test.shape)
```

```python
scaler = MinMaxScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
```

```python
X_train_scaled[:5]
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled[:5]
```

```python
# lr 
logreg = LogisticRegression(max_iter = 1000).fit(X_train_scaled, y_train)
print(logreg.score(X_test_scaled, y_test))
```

```python
# SVM 적용
from sklearn.svm import SVC
```

```python
svm = SVC(gamma = 'auto').fit(X_train_scaled, y_train)
print(svm.score(X_test_scaled, y_test))
```

```python
# KNN 적용
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1).fit(X_train_scaled, y_train)
print(knn.score(X_test_scaled, y_test))
```

```python
# pca
from sklearn.decomposition import PCA
pca = PCA(n_components = 10, whiten = True, random_state = 0).fit(X_train_scaled)
X_train_pca = pca.transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

knn = KNeighborsClassifier(n_neighbors = 1).fit(X_train_pca, y_train)
print(knn.score(X_test_pca, y_test))
```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter = 1000).fit(X_train_scaled, y_train)
print('train:', logreg.score(X_train_scaled, y_train))
print('test:', logreg.score(X_test_scaled, y_test))
```

```python
# PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
poly = PolynomialFeatures(degree = 5).fit(X_train_scaled)
X_train_poly = poly.transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

print(X_train_poly.shape, X_test_poly.shape)
```

```python
# 특성 자동 선택
from sklearn.feature_selection import SelectPercentile, f_classif
select = SelectPercentile(score_func=f_classif, percentile = 50).fit(X_train_scaled, y_train)
X_train_selected = select.transform(X_train_scaled)
X_test_selected = select.transform(X_test_scaled)
print(X_train_scaled.shape)
print(X_train_selected.shape)

lr = LogisticRegression(max_iter = 1000).fit(X_train_scaled, y_train)
print(lr.score(X_test_scaled, y_test))

lr.fit(X_train_selected, y_train)
print(lr.score(X_test_selected, y_test))
```

```python
# Grid Search
best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        svm = SVC(gamma = gamma, C = C).fit(X_train_l1, y_train)
        score = svm.score(X_test_l1, y_test)
        if score > best_score:
            best_score = score
            best_parameters = {'C' : C, 'gamma' : gamma}
            
print(best_score)
print(best_parameters)
```

```python
# 후진제거법 (변수 선택 5개 ~ 12개)
from sklearn.feature_selection import RFE

rf = RandomForestClassifier()

rfe_rf = RFE(rf, n_features_to_select = 5, step = 2)
model_rf = rfe_rf.fit(X_train_scaled, y_train)

print(model_rf.support_)

# model.ranking_
print(model_rf.score(X_test_scaled, y_test))
```
