```python
### 수식으로 표현 가능

X = range(-20, 21)
Y = [x**2+1 for x in X]
plt.plot(X, Y, 'o--')
plt.show()
```

```python
### 정수 범위 외의 그래프 그리기

import math as m                ## y = sin(2*pi*t)
X = range(0, 20)
Y = [m.sin(x) for x in X]
plt.plot(X, Y, 'o--')

## sin함수는 주기가 2pi , 
```

```python
import numpy as np
X = np.linspace(0, 10, 100)
Y = [m.sin(x) for x in X]
plt.plot(X, Y, 'o--')
plt.show()
```

```python
### sin 함수 그래프 그려보기  ( for 'np.linspace' ) 

X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)
plt.plot(X, Y)
plt.show()
```

```python
# SCATTER

dataA = np.random.standard_normal((100, 2))
dataB = np.random.standard_normal((100, 2))
dataC = np.random.standard_normal((100, 2))

dataA += np.array((-3, -3))
dataB += np.array((4, 4))
dataC += np.array((-1, 4))

plt.scatter(dataA[:, 0], dataA[:, 1], color = 'b')
plt.scatter(dataB[:, 0], dataA[:, 1], color = 'r')
plt.scatter(dataC[:, 0], dataA[:, 1], color = '0.6')            # 0 ~ 1 사이 값
plt.show()
```

```python
### EDGECOLOR OPTION

dataA = np.random.standard_normal((100, 2))
dataB = np.random.standard_normal((100, 2))
dataC = np.random.standard_normal((100, 2))

dataA += np.array((-3, -3))
dataB += np.array((4, 4))
dataC += np.array((-1, 4))

plt.scatter(dataA[:, 0], dataA[:, 1], edgecolor= '0', color = '#B778FF')
plt.scatter(dataB[:, 0], dataA[:, 1], edgecolor= '0', color = '#FFB2D9')
plt.scatter(dataC[:, 0], dataA[:, 1], edgecolor= '0', color = '0.6')
plt.show()
```

```python
# Color Map

import matplotlib.cm as cm
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X, Y, c=angle, cmap=cm.Spectral)           # cm.Greys,  cm.hsv ... 
plt.show()
```

```python
# APPLY TO LINE STYLE 

X = np.linspace(-10, 10, 100)
Y1 = [2*x**2+1 for x in X]
Y2 = [3*x**2+5 for x in X]
Y3 = [0.5*x**2+3 for x in X]

plt.plot(X, Y1, linestyle = 'solid')
plt.plot(X, Y2, linestyle = 'dashed')
plt.plot(X, Y3, linestyle = 'dashdot')

plt.show()
```

```python
# 누적막대그래프

N = 4
A = np.array([1, 2, 3, 2])
B = np.array([2, 3, 1, 2])
C = np.array([2, 3, 4, 5])
X = np.arange(N)
#plt.bar(X, A, color = 'r')
#plt.bar(X, B, color = 'b', bottom = A)
#plt.bar(X, C, color = 'k', bottom = A+B)
plt.bar(X, A, color = 'r')
plt.bar(X, A+B, color = 'b', bottom = A)
plt.bar(X, A+B+C, color = 'k', bottom = A+B)
plt.show()
```

```python
### HATCH OPTION
N = 4
A = np.array([1, 2, 3, 2])
B = np.array([2, 3, 1, 2])
C = np.array([2, 3, 4, 5])
X = np.arange(N)

plt.bar(X, A, color = 'r', hatch = '/')
plt.bar(X, A+B, color = 'b', hatch = 'x', bottom = A)
plt.bar(X, A+B+C, color = 'g', hatch = '/', bottom = A+B)
plt.show()
```

```python
### MARKER

A = np.random.standard_normal((100, 2))
B = np.random.standard_normal((100, 2))

A += np.array((-1, -2))
B += np.array((0.5, 1))
plt.scatter(A[:, 0], A[:, 1], color = 'k', marker = "D") # ^, o, x, v, <, >, s, p, P, X, D, ······
plt.scatter(B[:, 0], B[:, 1], color = 'k', marker = "*")
plt.show()
```

```python
### MARKEVERY OPTION

X = np.linspace(-6, 6, 1024)
Y1 = np.sinc(X)
Y2 = np.sinc(X)+1
plt.plot(X, Y1, marker='o', color='0.7')
plt.plot(X, Y2, marker='o', color='0.7', markevery = 32)            # markevery => '32개 점마다 하나씩' , 시계열 분석할때 용이
plt.show()
```

```python
### SIZE(S) OPTION ( BUBBLE CHART  =>  MAGNITUDE 
A = np.random.standard_normal((100, 2))
B = np.random.standard_normal((100, 2))

A += np.array((-1, -2))
B += np.array((0.5, 1))
plt.scatter(A[:, 0], A[:, 1], s = 10)
plt.scatter(B[:, 0], B[:, 1], s = 100)
plt.show()
```

```python
M = np.random.standard_normal((500, 2))
R = np.sum(M**2, axis = 1)
plt.figure(figsize=(5, 5))                                      # 사진 사이즈 
### x, y 축을 줄여서 범례, 라벨, 등등을 넣을 수 있는 공간 구획
plt.ylim(-5, 5)
plt.xlim(-5, 5)

plt.scatter(M[:, 0], M[:, 1], c = '0.7', marker = 's', s = 32*R, edgecolor = 'k')
plt.show()
```

```python
# 그래프 부분확대 효과

X = np.linspace(-6, 6, 1024)
X_small = np.linspace(-3, 3, 1024)
Y_small = np.sinc(X_small)
plt.plot(X, np.sinc(X), c = '0.7')
sub_axes = plt.axes([0.6, 0.6, 0.25, 0.25])
sub_axes.plot(X_small, Y_small, c = '0.4')
plt.show()
```

```python
# POLAR AXIS

T = np.linspace(0, 2*np.pi, 1024)
plt.axes(polar = True)                   # 극좌표계, POLAR
plt.plot(T, 1+0.25*np.sin(16*T))
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar=True)

radius = [1, 0.5, 0.7, 0.3, 0.1, 0.2, 0.6, 0.7]
theta = np.linspace(0, 2*np.pi, len(radius), endpoint = False)
points = np.vstack((theta, radius)).T
plt.gca().add_patch(patches.Polygon(points, color = '0.5'))
plt.show()
```
