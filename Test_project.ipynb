```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
hotel_df = pd.read_excel('C:/workspace/Python/hotel_bookings.xlsx')
```

```python
hotel_df.head()
```

```python
hotel_df.info()
```

```python
# 결측치 확인 - children, country, agent, company 결측치 존재.
hotel_df.isnull().sum()
```

```python
# 특성 삭제 - arrival_date_week_number, hotel, distribution_channel, company데이터 셋 삭제.
hotel_df = hotel_df.drop(['arrival_date_week_number', 'hotel', 'distribution_channel', 'company'], axis = 1)
```

```python
# 특성 대치 - children, agent, country
hotel_df['children'] = hotel_df['children'].fillna(0)
hotel_df['agent'] = hotel_df['agent'].fillna(-1)
hotel_df['country'] = hotel_df['country'].fillna('undefined')
```

```python
# 결측치 제거 완료.
hotel_df.info()
```

```python
# "arrival_date_month" 데이터를 숫자 형태로 바꾸기.

datetime_object = hotel_df['arrival_date_month'].str[0:3]
month_number = np.zeros(len(datetime_object))

for i in range(0, len(datetime_object)):
    datetime_object[i] = datetime.strptime(datetime_object[i], "%b")
    month_number[i] = datetime_object[i].month

# Float to integer conversion
month_number = pd.DataFrame(month_number).astype(int)
```

```python
# 'arrival date' 컬럼 생성(연-월-일 날짜 형식으로 변환한 컬럼.)
hotel_df['arrival_date'] = hotel_df['arrival_date_year'].map(str) + '-' + month_number[0].map(str) + '-' \
                       + hotel_df['arrival_date_day_of_month'].map(str)
```

```python
hotel_df.head()
```

```python
# 연, 월, 일 데이터 삭제.
hotel_df = hotel_df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1)
hotel_df.head()
```

```python
# arrival date 데이터 형식 조정
hotel_df['arrival_date'] = pd.to_datetime(hotel_df['arrival_date'])

print('Datatype of the arrival_date:', hotel_df['arrival_date'].dtype)
print('Datatype of the reservation_status_date:', hotel_df['reservation_status_date'].dtype)
```

```python
hotel_df['Total Guests'] = hotel_df['adults'] + hotel_df['children']
hotel_df['Total Guests'].head(10)
```

```python
hotel_df['Total Guests'].unique()
```

```python
# 전체 객수가 0명인 데이터 삭제.
hotel_df = hotel_df[hotel_df['Total Guests'] != 0]
```

```python
# 객수 0명 데이터 확인.
hotel_df['Total Guests'].unique()
```

```python
# 총 숙박일수 'Total Stays'
hotel_df['Total Stays'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']
```

```python
hotel_df.info()
```

```python
# 전처리된 데이터 저장.
hotel_df.to_csv("hotel_df_booking_predict.csv", index = False)
```
