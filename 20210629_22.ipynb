```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, random_state = 0)

# 데이터 스케일 조정
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

### MinMaxScaler를 사용해 스케일을 0에서 1 사이로 조정
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2).fit(X_train_scaled)
X_train_poly = poly.transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

print(X_train.shape)
print(X_test_poly.shape)

### 특성값(Feature)이 13개 --> 105개로 확장
```

```python
print(poly.get_feature_names())

### 원본 특성이 곱해져 새 특성이 만들어졌는지 정확환 관계를 보기 위해 get_feature_names() method 사용
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train_scaled, y_train)
print(ridge.score(X_test_scaled, y_test))   # 상호작용 특성이 없을때 점수
ridge = Ridge().fit(X_train_poly, y_train)
print(ridge.score(X_train_poly, y_train))   # 상호작용 특성이 있을때 점수

### 상호작용과 다항식 특성이 Ridge의 성능을 크게 높혀놓음.
```

```python
# RF
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100, random_state = 0).fit(X_train_scaled, y_train)
print(rf.score(X_test_scaled, y_test))   # 상호작용 특성이 없을때 점수
rf = RandomForestRegressor(n_estimators = 100, random_state = 0).fit(X_train_poly, y_train)
print(rf.score(X_test_poly, y_test))    # 상호작용 특성이 있을때 점수

### 특성을 추가안해도 RF는 Ridge의 성능과 맞먹음, 오히려 상호작용과 다항식을 추가하면 성능 down
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

# 고정된 난수를 발생시킨다
rng = np.random.RandomState(42)
noise = rng.normal(size=(len(cancer.data), 50))

# 데이터에 노이즈 특성 추가
X_w_noise = np.hstack([cancer.data, noise])

X_train, X_test, y_train, y_test = train_test_split(X_w_noise, cancer.target, random_state = 0, test_size = .5)

# f_classif(기본값)와 SelecPercentile을 사용하여 특성의 50%를 선택.
select = SelectPercentile(score_func = f_classif, percentile = 50).fit(X_train, y_train)

# 훈련 세트에 적용
X_train_selected = select.transform(X_train)

print(X_train.shape)
print(X_train_selected.shape)


### 특성개수(feature)가 80개 --> 40개로 감소(원본 특성의 50%)
```

```python
# get_support 메서드는 선택된 특성을 불리언 값으로 표시해주어 어떤 특성이 선택됐는지 확인해준다.
mask = select.get_support()
print(mask)


plt.matshow(mask.reshape(1, -1), cmap = 'gray_r')
plt.xlabel('특성 번호')
plt.yticks([0])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# 테스트 데이터 변환
X_test_selected = select.transform(X_test)

lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
print(lr.score(X_test, y_test))            # 전체 특성을 사용한 점수
lr.fit(X_train_selected, y_train)
print(lr.score(X_test_selected, y_test))   # 선택된 일부 특성을 사용한 점수
```

```python
citibike = mglearn.datasets.load_citibike()
citibike.head()
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font) 

plt.figure(figsize=(10, 3))
xticks=pd.date_range(start = citibike.index.min(), 
                     end = citibike.index.max(), freq = 'D')

week = ['일', '월', '화', '수', '목', '금', '토']
xticks_name = [week[int(w)] + d for w, d in zip(xticks.strftime('%w'),
                                               xticks.strftime(' %m-%d'))]
plt.xticks(xticks, xticks_name, rotation = 90, ha = 'left')
plt.plot(citibike, marker = 'x')
plt.show()
```

```python
# 시계열 데이터에 있어서 훈련데이터 선택할땐, 현재와 가장 가까운 데이터를 훈련데이터로 선정한다. (08.24 ~ 08.31)
# 훈련 (08.01 ~ 08.23) / 테스트 (08.24 ~ 08.31)

# 타깃 값 추출 (대여횟수)
y = citibike.values
# POSIX 시간을 10**9로 나누어 변환
X = citibike.index.astype('int64').values.reshape(-1, 1) // 10**9
X
```

```python
# 처음 184개 데이터 포인터를 훈련 세트로 사용하고 나머지는 테스트로 사용
n_train = 8 * 23   ## 184개

# 주어진 특성을 사용해 평가하고 그래프를 만듦
def eval_on_features(features, target, regressor):
    # 훈련 셋 / 테스트 셋 나누기
    X_train, X_test = features[:n_train], features[n_train:]
    # 타깃 값도 나누기
    y_train, y_test = target[:n_train], target[n_train:]
    regressor.fit(X_train, y_train)
    print(regressor.score(X_test, y_test))
    y_pred = regressor.predict(X_test)
    y_pred_train = regressor.predict(X_train)
    
    
    plt.figure(figsize=(10, 3))
    plt.xticks(range(0, len(X), 8), xticks_name, rotation = 90, ha='left')
    
    plt.plot(range(n_train), y_train, label = '훈련')
    plt.plot(range(n_train, len(y_test) + n_train), y_test, '-', label = '테스트')
    plt.plot(range(n_train), y_pred_train, '--', label = '훈련 예측')
    plt.plot(range(n_train, len(y_test) + n_train), y_pred, '--', label = '테스트 예측')
    
    plt.legend(loc=(1.01, 0))
    plt.xlabel('날짜')
    plt.ylabel('대여횟수')
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
eval_on_features(X, y, regressor)


### RF 훈련 셋은 아주 훌륭 , 그러나 테스트는 한 가지 값으로만 에측함.
### R^2 값은 -0.04로 거의 아무것도 학습하지 못했음. 이유는????????

### => RF와 특성 사이의 조합. Test set에 있는 POSIX 시간 특성의 값은 Train set에 있는 특성 값의 범위 밖에 있다. 
###    즉 테스트 세트에 있는 데이터 포인트는 훈련 세트에 있는 모든 데이터보다 뒤의 시간이다. 
###    트리 모델인 RF는 훈련 세트에 있는 특성의 범위 밖으로 외샵(extrapolation) 할 수 있는 능력이 없다. 
###    결국, 이 모델은 테스트 세트와 가장 가까이 있는 마지막 훈련 세트데이터의 타깃 갑을 에측으로 사용하게 된다.


### 문제점 : 시간과 요일이 반영 안된듯, 이를 직접 추가. POSIX 시간으로는 아무것도 학습되지 않으므로 이 특성은 제외.

### 해결책 : 전문가 지식이 필요, 시간과 요일 특성을 추가.
```
